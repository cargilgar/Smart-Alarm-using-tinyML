Serie takes a Python list a an input. It only has 1 dimension

# Manipuating data

When working with data, our dataframes are going to be missing data in some cells as well as having columns with undesired format. We need to be able to tackle this problem

```python
# Importing car_sales detaframe with missing data:

car_sales_missing = pd.read_csv("/home/carlosg/dev/zero-to-mastery-ml-master/data/car-sales-missing-data.csv")
car_sales_missing
```

## Standard types: NaN, NA or empty cells

These are standard missing values that Pandas can easily handle

### Filling NaN cells (Not a Number)

```python
# Filling NaN values in Odometer with the mean value of Odometer (for example)
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

#### Note: 
Although the NaN values seem to have dissapeared, they have not been replaced  in the DataFrame.
To apply the changes we need to reassign the value:

```python
# car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

However, there is a more elegant way of doing this. By using 'inplace' boolean:

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)
car_sales_missing
```

### Deleting any row that has missing values

```python
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

## Non-standard types: na, n/a, 

These are usually typed manually and not following any convention

This can be solved easily by creating a list and importing the file again. Pandas will then detect the new missing values:

# Adding a new column to an exisiting DataFrame

## New column from Series

```python
seats_column = pd.Series([5, 5, 5, 5, 5])

car_sales["Seats"] = seats_column  # Adding new column with label "Seats" to the DataFrame
car_sales
```

Note: by default it will start filling from the first index, leaving the rest (if there are more) unfilled.
We will need to handle missing values afterwards

## New column from a Python list (not recommendable)

```python
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 3.6, 7.8, 5.6, 9.8, 8.2]

car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

Note: unlike Series, the python list has to be the same length as the DataFrame. Otherwise, it will throw an error.

## New column from existing columns

```python
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

## New column from a value

```python
car_sales["Number of wheels"] = 4
car_sales
```

# Deleting a column

Note: to specify whether a row or column:

Rows: axis=0

Columns: axis=1

```python
car_sales = car_sales.drop("Number of wheels", axis=1)  # Remember to reassign
car_sales
```

## Shuffling the data
Useful for randomising our data into the machine learning model

```python
car_sales_shuffled = car_sales.sample(frac=1.0)  # Shuffing 100% of the data
car_sales_shuffled
```

## Getting shuffled data back in order

```python
car_sales_shuffled.reset_index(drop=True, inplace=True)  # drop=True avoids adding a new index column
car_sales_shuffled
```

## Apply a function to a certain column
It can be lambda, numpy...

```python
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x/1.6)  # Converting KM to Miles
car_sales
```

Note: after changing the units, we need to rename the column from KM to Miles

## Rename a column

```python
car_sales = car_sales.rename({"Odometer (KM)": "Odometer (Miles)"}, axis=1)
car_sales
```
